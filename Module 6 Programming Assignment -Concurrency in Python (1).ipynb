13.1 Write current date to today.txt

```python
from datetime import datetime

# Get current date as string
today = datetime.now().strftime("%Y-%m-%d")

# Write to file
with open("today.txt", "w") as file:
    file.write(today)

print("Date written to today.txt:", today)
```

13.2 Read today.txt into today_string

```python
# Read file into string
with open("today.txt", "r") as file:
    today_string = file.read()

print("Read from file:", today_string)
```

13.3 Parsse the date from today_string

```python
from datetime import datetime

# Convert string back to date object
parsed_date = datetime.strptime(today_string, "%Y-%m-%d")

print("Parsed date:", parsed_date)
```

Create 3 processes that wait random time and print time

```python
import multiprocessing
import time
import random
from datetime import datetime

# Function for each process
def worker():
    wait_time = random.random()
    time.sleep(wait_time)
    print("Process:", multiprocessing.current_process().name)
    print("Current Time:", datetime.now().strftime("%H:%M:%S"))
    print("Waited:", wait_time, "seconds\n")

if __name__ == "__main__":
    processes = []

    # Create 3 processes
    for i in range(3):
        p = multiprocessing.Process(target=worker)
        processes.append(p)
        p.start()

    # Wait for all to finish
    for p in processes:
        p.join()

    print("All processes finished.")
```
